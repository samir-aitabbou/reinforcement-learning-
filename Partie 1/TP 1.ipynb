```python
def alpha(c,prob):
            sommeProb=0
            alpha=0
            for i in range(len(prob)):
                sommeProb += prob[i]
                # if channel capacity is reaches 
                if sommeProb <= c:
                    alpha+=1
                    
                else:
                    alpha += (c- (sommeProb-prob[i])) / prob[i]
                    break
                
            return alpha 

p=[1/3,1/12, 1/4, 1/3]
alpha=alpha(0.34,p)
alpha
```

```python
def kns(rewards, prob, c):
    n = len(rewards)
    packets = list(range(1, n+1)) # on aura 4 packets: 1, 2, 3 et 4
    packets.sort(key=lambda i: rewards[i-1], reverse=True)
    sommePoid = 0   # summ des poids avant de dpasse la capacit du seuil
    policy = [0] * n #initialise la politique en liste 
    #print(policy)
    for packet in packets:
        if sommePoid + prob[packet-1] <= c:
            policy[packet-1] = 1
            sommePoid += prob[packet-1]
        else:
            policy[packet-1] = (c - sommePoid) / prob[packet-1]
            break
    return policy

rewards = [100, 50, 10, 1]
prob = [1/3, 1/12, 1/4, 1/3]
c = 0.34
policy = kns(rewards, prob, c)
print("Optimal policy:", policy)
```

```python
import random
import numpy as np
import collections
import matplotlib.pyplot as plt


N = 4
lamb = 10**5
c = 0.34 * lamb
#c = 0.34
r = [100, 50, 10, 1]
p = [1/3, 1/12, 1/4, 1/3]
num_episodes = 1000
alpha = 2.6 # alpha_opti = 1.08

#  rpter les indices [0, 1, 2, 3] avec des frquences spcifiques dtermines par les produits des probabilits p et de lamb
paquet = np.repeat([0, 1, 2, 3], [p[0]*lamb, p[1]*lamb, p[2]*lamb, p[3]*lamb])

#  nombre d'occurrences de chaque indice dans le tableau paquet
nb_paquet = [collections.Counter(paquet)[0], collections.Counter(paquet)[1], collections.Counter(paquet)[2], collections.Counter(paquet)[3]]


def estimate_Y(admitted, refused):
    denominator = admitted + refused
    if denominator == 0:
        return 0  # Avoid division by zero
    return admitted / (denominator)

def policy(alpha_n):

    remain_capacity = c
    total_accepted = 0
    total_refused = 0
    u = [0,0,0,0]

    for i in range(0, lamb-1):
        if paquet[i] <= int(alpha_n) and u[paquet[i]] < (alpha_n - paquet[i])*nb_paquet[paquet[i]] and remain_capacity > 0:
            u[paquet[i]] += 1
            total_accepted += 1
            remain_capacity -= 1
        else:
            total_refused += 1
    return u, total_accepted, total_refused


def robbins_monro(epsilon, gamma):
    all_alpha = []
    alpha_n = alpha
    alpha_average = alpha
    for episode in range(num_episodes):
        np.random.shuffle(paquet)
        epsilon_n = epsilon/((episode+1)**gamma)

        #print(epsilon_n)

        u, total_accepted, total_refused = policy(alpha_n)   # Adopt policy
        #u, total_accepted, total_refused = policy(alpha_average)
        print(u)
        Yn = estimate_Y(total_accepted, total_refused)   # estimate Yn

        #print(Yn)
        #alpha_average = (1 - 1 / (episode + 1)) * alpha_average + (1 / (episode + 1)) * alpha_n  # Mise  jour des Polyak Averages
        alpha_n = max(0, min(alpha_n + epsilon_n * (c - Yn*lamb), N))
        all_alpha.append(alpha_n)
        """
        if len(all_alpha) % 5 == 0:
            alpha_average = 0
            for alph in all_alpha[-5:]:
                alpha_average += alph
            alpha_average /= 5
        all_alpha.append(alpha_average)  # Utilisation de la moyenne d'alpha
        """


        #print(alpha_n)
    return alpha_n, all_alpha

# decreasing stepsize
alpha_etoile, all_alpha = robbins_monro(epsilon=1/N, gamma=0.7)

#alpha_etoile, all_alpha = robbins_monro(epsilon=0.005, gamma=0.7)

# constant stepsize
#alpha_etoile, all_alpha = robbins_monro(epsilon=0.005, gamma=0)


plt.plot(all_alpha)


plt.title("Graphique de valeurs")
plt.xlabel("Nombre d'itration")
plt.ylabel("alpha")

# Affichez le graphique
plt.show()

#print(robbins_monro_decreasing_stepsize())

```
