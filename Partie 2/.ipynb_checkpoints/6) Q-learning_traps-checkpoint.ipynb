```python
import numpy as np
import matplotlib.pyplot as plt
import time

class TrapsGridWorld:
    def __init__(self, rows=12, cols=12):
        self.start = (0, 0)
        self.goal = (0, 0)
        self.rows = rows
        self.cols = cols
        self.traps = [(0, 3), (1, 3), (2, 3), (3, 3), (4, 3), (5, 3), (6, 3), (7, 3),
                      (4, 7), (5, 7), (6, 7), (7, 7), (8, 7), (9, 7), (10, 7), (11, 7)]
        self.actions_list = ['N', 'E', 'S', 'W']
        self.Q_table = {(x, y): {act: 0 for act in self.actions_list} for x in range(cols) for y in range(rows)}

    def set_terminal(self, start_state, goal_state):
        self.start = start_state
        self.goal = goal_state

    def next_state(self, current_state, action):
        x, y = current_state
        del_x, del_y = 0, 0
        if action == 'E':
            del_x = 1
        elif action == 'W':
            del_x = -1
        elif action == 'N':
            del_y = -1
        elif action == 'S':
            del_y = 1
        else:
            raise ValueError('Invalid action!')
        new_x = max(0, min(x + del_x, self.cols - 1))
        new_y = max(0, min(y + del_y, self.rows - 1))
        if (new_x,new_y) in self.traps:
            r = self.reward_function((new_x,new_y))
            return r,current_state
        else:
            r = self.reward_function((new_x,new_y))
            return r,(new_x,new_y)

    def check_terminal(self, state):
        return state == self.goal

    def reward_function(self, state_prime):
        if state_prime == self.goal:
            return 2 * (self.cols - 1)  
        elif state_prime in self.traps:
            return -2 * (self.cols - 1)
        else:
            return -1
def eps_greedy(episode, q_dict, epsilon, actions):
    def greedy_act(q_dict):
        max_q = max(q_dict.values())
        return [act for act, val in q_dict.items() if val == max_q]

    m = len(actions)
    greedy_actions = greedy_act(q_dict)
    p = [(epsilon * 1. / m) + (1 - epsilon) if act in greedy_actions else epsilon * 1. / m for act in actions]
    p = np.array(p)
    p /= np.sum(p)
    choice = np.random.choice(actions, size=1, p=p)
    return choice[0]



def grid_world_q_learning(world, start_state, goal_state, alpha=0.1, gamma=1, epsilon=0.1):
    world.set_terminal(start_state, goal_state)
    
    step_ep_list = []
    ep, ep_max = 1, 300
    step = 0

    while ep < ep_max:
        state = start_state
        trajectory = [state]

        while not world.check_terminal(state):
            state_coords = state
            act = eps_greedy(ep, world.Q_table[state_coords], epsilon, world.actions_list)
            reward,next_state = world.next_state(state_coords, act)
            #reward = world.reward_function(next_state)
            # Modifier l'équation de mise à jour pour Q-learning
            max_next_q = max(world.Q_table[next_state].values())
            world.Q_table[state_coords][act] += alpha * (reward + gamma * max_next_q - world.Q_table[state_coords][act])
            state = next_state
            step += 1
            step_ep_list.append(ep)
            trajectory.append(state)

        if ep == (ep_max - 1):
            trajectory = trajectory_path(world, trajectory)

        ep += 1

    plt.plot(step_ep_list)
    plt.title('GridWorld_Q-learning ', fontsize='large')
    plt.xlabel("Number of Steps taken")
    plt.ylabel("Number of Episodes")
    plt.show()

    return trajectory

def trajectory_path(world, traj):
    world_map = np.zeros((world.rows, world.cols))
    for i, state in enumerate(traj):
        x, y = state
        world_map[y, x] = i + 1
    print(world_map)
    print("\n")
    return world_map

start_state = (0, 0)
goal_state = (11, 11)
world = TrapsGridWorld()

print("Plot for Option: ε=0.1, α=0.5 \n")
traj = grid_world_q_learning(world, start_state, goal_state, alpha=0.5, epsilon=0.1)

def display_policy(world):
    optimal_policy = np.zeros((world.rows, world.cols), dtype=str)
    for y in range(world.rows):
        for x in range(world.cols):
            state = (x, y)
            if state == world.goal:
                optimal_policy[y, x] = 'G'
                continue
            best_action = max(world.Q_table[state], key=world.Q_table[state].get)
            optimal_policy[y, x] = best_action
    print("Optimal Policy:")
    print(optimal_policy)

# Afficher la fonction de valeur
def display_value_function(world):
    value_function = np.zeros((world.rows, world.cols))
    for y in range(world.rows):
        for x in range(world.cols):
            state = (x, y)
            value_function[y, x] = max(world.Q_table[state].values())
    print("\nValue Function:")
    print(value_function)
display_policy(world)
display_value_function(world)
```

```python

import numpy as np
import matplotlib.pyplot as plt
traps = [(3, 0), (3, 1), (3, 2), (3, 3), (3, 4), (3, 5), (3, 6), (3, 7),
                      (7, 4), (7, 5), (7, 6), (7, 7), (7, 8), (7, 9), (7,10), (7, 11)]
def plot_path(matrix, traps):
    matrix_size = np.shape(matrix)
    fig, ax = plt.subplots(figsize=matrix_size)

    for i in range(matrix_size[0]):
        for j in range(matrix_size[1]):
            if matrix[i, j] == 0 and (i,j) in traps:
                circle = plt.Circle((j, i), 0.2, color='red')
                ax.add_artist(circle)
                ax.text(j, i, 'TRAP', ha='center', va='center', color='white', fontsize=8)
            elif matrix[i, j] == 0:
                circle = plt.Circle((j, i), 0.2, color='blue')
                ax.add_artist(circle)
                ax.text(j, i, 'STOP', ha='center', va='center', color='white', fontsize=8)
            elif matrix[i, j] != 0 and (i==11 and j==11):
                circle = plt.Circle((j, i), 0.2, color='green')
                ax.add_artist(circle)
                ax.text(j, i, 'GOAL', ha='center', va='center', color='white', fontsize=8)
            else:
                circle = plt.Circle((j, i), 0.2, color='green')
                ax.add_artist(circle)
                ax.text(j, i, 'GO', ha='center', va='center', color='white', fontsize=8)

    ax.set_xlim(-0.5, matrix_size[1] - 0.5)
    ax.set_ylim(-0.5, matrix_size[0] - 0.5)
    ax.set_aspect('equal')
    ax.set_xticks(np.arange(matrix_size[1]))
    ax.set_yticks(np.arange(matrix_size[0]))
    ax.set_xticklabels(np.arange(matrix_size[0]))
    ax.set_yticklabels(np.arange(matrix_size[0]))
    ax.grid(which='both', color='black', linestyle='-', linewidth=1)
    plt.show()



plot_path(traj, traps)
```
