```python
import numpy as np
import matplotlib.pyplot as plt
import time
import random

class TrapsGridWorld:
    def __init__(self, rows=12, cols=12):
        self.start = (0, 0)
        self.goal = (11, 11)
        self.rows = rows
        self.cols = cols
        self.traps = [(0, 3), (1, 3), (2, 3), (3, 3), (4, 3), (5, 3), (6, 3), (7, 3),
                      (4, 7), (5, 7), (6, 7), (7, 7), (8, 7), (9, 7), (10, 7), (11, 7)]
        
        self.actions_list = ['N', 'E', 'S', 'W']
        #self.Q_table = {(x, y): {act:0 for act in self.actions_list} for x in range(cols) for y in range(rows)}
        self.Q_table = {
            (x, y): {act: random.random() if (x, y) != (cols - 1, rows - 1) else 0 for act in self.actions_list}
            for x in range(cols)
            for y in range(rows)
        }

    def set_terminal(self, start_state, goal_state):
        self.start = start_state
        self.goal = goal_state

    def next_state(self, current_state, action):
        x, y = current_state
        del_x, del_y = 0, 0
        if action == 'E':
            del_x = 1
        elif action == 'W':
            del_x = -1
        elif action == 'N':
            del_y = -1
        elif action == 'S':
            del_y = 1
        else:
            raise ValueError('Invalid action!')
        new_x = max(0, min(x + del_x, self.cols - 1))
        new_y = max(0, min(y + del_y, self.rows - 1))
        if (new_x,new_y) in self.traps:
            r = self.reward_function((new_x,new_y))
            return r,current_state
        else:
            r = self.reward_function((new_x,new_y))
            return r,(new_x,new_y)
    

    def check_terminal(self, state):
        return state == self.goal

    def reward_function(self, state_prime):
        if state_prime == self.goal:
            return 2 * (self.cols - 1)  
        elif state_prime in self.traps:
            return -2 * (self.cols - 1)
        else:
            return -1
def eps_greedy(q_dict, epsilon, actions):
    def greedy_act(q_dict):
        max_q = max(q_dict.values())
        return [act for act, val in q_dict.items() if val == max_q]

    m = len(actions)
    greedy_actions = greedy_act(q_dict)
    p = [(1 - epsilon) if act in greedy_actions else epsilon for act in actions]
    p = np.array(p)
    p /= np.sum(p)
    choice = np.random.choice(actions, size=1, p=p)
    return choice[0]

def grid_world_sarsa(world, start_state, goal_state, alpha=0.1, gamma=0.9, epsilon=0.1):
    world.set_terminal(start_state, goal_state)
    
    step_ep_list = []
    ep, ep_max = 1, 300
    step = 0

    while ep < ep_max:
        state = start_state
        trajectory = [state]

        while not world.check_terminal(state):
            current_state = state
            act = eps_greedy(world.Q_table[current_state], epsilon, world.actions_list)
            reward,next_state = world.next_state(current_state, act)
            #reward = world.reward_function(next_state)
            next_act = eps_greedy(world.Q_table[next_state], epsilon, world.actions_list)
            world.Q_table[current_state][act] += alpha * (reward + gamma * world.Q_table[next_state][next_act] - world.Q_table[current_state][act])
            state = next_state
            step += 1
            step_ep_list.append(ep)
            trajectory.append(state)

        if ep == (ep_max - 1):
            trajectory = trajectory_path(world, trajectory)

        ep += 1

    plt.plot(step_ep_list)
    plt.title('GridWorld_SARSA ', fontsize='large')
    plt.xlabel("Number of Steps taken")
    plt.ylabel("Number of Episodes")
    plt.show()

    return trajectory 

def trajectory_path(world, traj):
    world_map = np.zeros((world.rows, world.cols))
    for i, state in enumerate(traj):
        x, y = state
        world_map[y, x] = i + 1
    print(world_map)
    print("\n")
    return world_map

start_state = (0, 0)
goal_state = (11, 11)
world = TrapsGridWorld()

print("Plot for Option: ε=0.1, α=0.5 \n")
traj = grid_world_sarsa(world, start_state, goal_state, alpha=0.5, epsilon=0.1)

```

```python
def display_policy(world):
    optimal_policy = np.zeros((world.rows, world.cols), dtype=str)
    for y in range(world.rows):
        for x in range(world.cols):
            state = (x, y)
            if state == world.goal:
                optimal_policy[y, x] = 'G'
                continue
            best_action = max(world.Q_table[state], key=world.Q_table[state].get)
            optimal_policy[y, x] = best_action
    print("Optimal Policy:")
    print(optimal_policy)

# Afficher la fonction de valeur
def display_value_function(world):
    value_function = np.zeros((world.rows, world.cols))
    for y in range(world.rows):
        for x in range(world.cols):
            state = (x, y)
            value_function[y, x] = max(world.Q_table[state].values())
    print("\nValue Function:")
    print(value_function)
display_policy(world)
display_value_function(world)
```

```python

import numpy as np
import matplotlib.pyplot as plt
traps = [(3, 0), (3, 1), (3, 2), (3, 3), (3, 4), (3, 5), (3, 6), (3, 7),
                      (7, 4), (7, 5), (7, 6), (7, 7), (7, 8), (7, 9), (7,10), (7, 11)]
def plot_path(matrix, traps):
    matrix_size = np.shape(matrix)
    fig, ax = plt.subplots(figsize=matrix_size)

    for i in range(matrix_size[0]):
        for j in range(matrix_size[1]):
            if matrix[i, j] == 0 and (i,j) in traps:
                circle = plt.Circle((j, i), 0.2, color='red')
                ax.add_artist(circle)
                ax.text(j, i, 'TRAP', ha='center', va='center', color='white', fontsize=8)
            elif matrix[i, j] == 0:
                circle = plt.Circle((j, i), 0.2, color='blue')
                ax.add_artist(circle)
                ax.text(j, i, 'STOP', ha='center', va='center', color='white', fontsize=8)
            elif matrix[i, j] != 0 and (i==11 and j==11):
                circle = plt.Circle((j, i), 0.2, color='green')
                ax.add_artist(circle)
                ax.text(j, i, 'GOAL', ha='center', va='center', color='white', fontsize=8)
            else:
                circle = plt.Circle((j, i), 0.2, color='green')
                ax.add_artist(circle)
                ax.text(j, i, 'GO', ha='center', va='center', color='white', fontsize=8)

    ax.set_xlim(-0.5, matrix_size[1] - 0.5)
    ax.set_ylim(-0.5, matrix_size[0] - 0.5)
    ax.set_aspect('equal')
    ax.set_xticks(np.arange(matrix_size[1]))
    ax.set_yticks(np.arange(matrix_size[0]))
    ax.set_xticklabels(np.arange(matrix_size[0]))
    ax.set_yticklabels(np.arange(matrix_size[0]))
    ax.grid(which='both', color='black', linestyle='-', linewidth=1)
    plt.show()



plot_path(traj, traps)
```

```python
import argparse
import numpy as np
import random

import matplotlib.pyplot as plt
from matplotlib import colors


class Environment:
    def __init__(self, n, actions, rewards, e_x, e_y):
        self.n = n
        self.actions = actions
        self.rewards = rewards
        self.e_x = e_x
        self.e_y = e_y


def sarsa(n, p_barrier, r_barrier, v0_val, gamma, epsilon, alpha, max_steps, seed_nr):
    """Runs a simulation of the SARSA algorithm.

    Args:
        n (int): length and width of the grid
        p_barrier (float): probability of a cell being a barrier
        r_barrier (int): reward for the barrier cells
        v0_val (int): initial value for the value function
        gamma (float): gamma parameter (between 0 and 1)
        epsilon (float): exploration-exploitation trade-off parameter
        alpha (float): learning rate
        max_steps (int): maximum number of steps to run the algorithm
        seed_nr (int): seed number (for reproducible results)
    """

    # Create initial environment
    env = build_grid(n, p_barrier, r_barrier, seed_nr)
    i = 0
    plot_grid(env)

    # Generate initial Q-values
    Q = get_init_Q(n, v0_val, env.e_x, env.e_y)

    # Plot initial Q-values
    plot_q_values(Q, env.n)

    for _ in range(max_steps):
        i += 1
        print("\nStep {} of SARSA algorithm".format(i))

        # Run SARSA update rule
        sarsa_update(env, Q, gamma, epsilon, alpha)
        
        # Plot Q-values
        plot_q_values(Q, env.n)

    print("\nSARSA algorithm completed after {} steps".format(i))


def sarsa_update(env, Q, gamma, epsilon, alpha):
    """Applies the SARSA update rule.

    Args:
        env (Environment): gridworld environment
        Q (array): numpy array representing the Q-values
        gamma (float): gamma parameter (between 0 and 1)
        epsilon (float): exploration-exploitation trade-off parameter
        alpha (float): learning rate
    """

    # Initialize state
    x, y = 0, 0
    state = (x, y)
    action = epsilon_greedy_policy(Q, state, env.actions, epsilon)

    # Run one episode
    while not is_terminal_state(state, env):
        # Take action and observe next state and reward
        next_x, next_y = get_next_state(x, y, action, env.n)
        next_state = (next_x, next_y)
        reward = env.rewards[next_x, next_y]

        # Choose next action using epsilon-greedy policy
        next_action = epsilon_greedy_policy(Q, next_state, env.actions, epsilon)

        # Update Q-value using SARSA update rule
        Q[x, y, action] += alpha * (reward + gamma * Q[next_x, next_y, next_action] - Q[x, y, action])

        # Move to the next state and action
        x, y = next_x, next_y
        state = next_state
        action = next_action


def epsilon_greedy_policy(Q, state, actions, epsilon):
    """Epsilon-greedy policy for action selection.

    Args:
        Q (array): numpy array representing the Q-values
        state (tuple): current state (x, y)
        actions (list): list of possible actions
        epsilon (float): exploration-exploitation trade-off parameter

    Returns:
        action (int): selected action
    """

    if random.uniform(0, 1) < epsilon:
        # Exploration: choose a random action
        return random.choice(actions)
    else:
        # Exploitation: choose the action with the highest Q-value
        return np.argmax(Q[state[0], state[1], :])


def is_terminal_state(state, env):
    """Check if the given state is a terminal state.

    Args:
        state (tuple): current state (x, y)
        env (Environment): gridworld environment

    Returns:
        is_terminal (bool): True if the state is a terminal state, False otherwise
    """
    return state[0] == env.e_x and state[1] == env.e_y


def get_init_Q(n, v0, e_x, e_y):
    """Initialize Q-values.

    Args:
        n (int): length and width of the grid
        v0 (float): initial value for the Q-values
        e_x (int): x value of the end cell
        e_y (int): y value of the end cell

    Returns:
        Q (array): initial Q-values
    """

    Q = v0 * np.ones((n, n, 4))  # 4 actions
    # Q-values for terminal state are 0
    Q[e_x, e_y, :] = 0
    return Q


def plot_q_values(Q, n):
    """Plots the Q-values in each state as a grid.

    Args:
        Q (array): numpy array representing the Q-values
        n (int): length and width of the grid
    """

    fig, ax = plt.subplots()
    max_Q = np.max(Q)
    min_Q = np.min(Q)

    # create discrete colormap
    cmap = colors.ListedColormap(['white', 'orange', 'red', 'purple', 'blue'])
    bounds = [min_Q, 0, max_Q / 2, max_Q]
    norm = colors.BoundaryNorm(bounds, cmap.N)

    im = ax.imshow(np.max(Q, axis=-1), cmap=cmap, norm=norm)

    # draw gridlines
    ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=2)
    ax.set_xticks(np.arange(-.5, n, 1))
    ax.set_yticks(np.arange(-.5, n, 1))

    # Loop over data dimensions and create text annotations.
    for i in range(n):
        for j in range(n):
            for a in range(4):
                text = ax.text(j, i, "{:.2f}".format(Q[i, j, a]), ha="center", va="center", color="black")

    ax.set_title("Q-values")
    fig.tight_layout()
    plt.show()


if __name__ == "__main__":
    # Specify arguments
    args_list = [
        "--n", "12",
        "--p_barrier", "0.1",
        "--r_barrier",  str(-2 * (12 - 1)),
        "--v0_val", "0",
        "--gamma", "0.9",
        "--epsilon", "0.1",
        "--alpha", "0.1",
        "--max_steps", "100"
#         "--seed_nr
```
